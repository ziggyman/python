```python
from astropy.io import fits
import numpy as np
import os
from pyraf import iraf

iraf.stsdas()
iraf.task(lacos_im = '/Volumes/obiwan/Users/azuri/stella/lacos_im.cl')

gain = 0.95
rdnoise = 4.5
sigclips = [2.5,3.0,3.5,4.0,4.5,5.0]
sigfracs = [0.3,0.4,0.5,0.6,0.7]
objlims = [0.5,0.75,1.0,1.25,1.5]
niters = [1,2,3,4,5]

directory = '/Volumes/obiwan/azuri/spectra/IPHAS_GTC/Laurence/GTC4-16AMEX/OB0065/'
inFile = os.path.join(directory,'gtc_object_av.fits')
datFile = os.path.join(directory,'gtc_object_av_cr.fits')
outFileRoot = os.path.join(directory,'gtc_object_av_cr_lacos_im_c%.1f_f%.1f_l%.1f_i%1d.fits')
outMaskRoot = os.path.join(directory,'gtc_object_av_cr_lacos_im_c%.1f_f%.1f_l%.1f_i%1d.pl')
dataSec = '[1200:1600,1:2051]'
if os.path.isfile(datFile):
    iraf.imdel(datFile)
iraf.imcopy(input=inFile+dataSec, output=datFile, verbose=False)

for clip in sigclips:
    for frac in sigfracs:
        for lim in objlims:
            for niter in niters:
                print 'running clip = ',clip,', frac = ',frac,', lim = ',lim,', niter = ',niter
                iraf.lacos_im(input=datFile,
                              output=outFileRoot % (clip, frac, lim, niter),
                              outmask=outMaskRoot % (clip, frac, lim, niter),
                              gain=gain,
                              readn=rdnoise,
                              statsec='*,*',
                              skyval=0.,
                              sigclip=clip,
                              sigfrac=frac,
                              objlim=lim,
                              niter=niter)
```
